```python
turl = "https://min-api.cryptocompare.com/data/histoday?fsym=USDT&tsym=USD&e=Kraken&limit=184&toTs=1562101200"
usdt = requests.get(turl + "&api_key=" + api_key)
usdtdat = usdt.json()
usdtdattab = pd.DataFrame(usdtdat['Data'])
usdtdattab
```

```python
usdtcleaned=usdtdattab[['open', 'time', 'close']]
usdtcleaned=usdtcleaned.rename(columns={'open':'tether_open', 'close':'tether_close'})
btccleaned=btcdattab[['open', 'time', 'close']]
btccleaned=btccleaned.rename(columns={'open':'btc_open', 'close': 'btc_close'})
```

```python
combined = pd.merge(usdtcleaned, btccleaned, on='time', how='outer')
combined
```

```python
combined['time'] = pd.to_datetime(combined['time'], unit = 's')
combined
```

```python
combined['tether_peg']= ((combined['tether_open']+combined['tether_close'])/2) - 1
combined['tether_change']= 100*((combined['tether_close']-combined['tether_open'])/((combined['tether_open']+combined['tether_close'])/2))
combined['btc_change']= 100*((combined['btc_close']-combined['btc_open'])/((combined['btc_open']+combined['btc_close'])/2))
combined['btc change/tether change'] = combined['btc_change']/combined['tether_change']
combined = combined.set_index(['time'], drop = False)
combined
```

```python
fig, ax1 = plt.pyplot.subplots()

color = 'tab:red'
ax1.set_xlabel('Time')
ax1.set_ylabel('Change in BTC price',  color = color)
ax1.plot(combined['time'], combined['btc_change'], color=color)
ax1.tick_params(axis='y', labelcolor='red')

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('Change in Tether Price', color=color)  # we already handled the x-label with ax1
ax2.plot(combined['time'], combined['tether_change'], color=color, )
ax2.tick_params(axis='y', labelcolor=color)
fig.set_size_inches(30,24)
fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.pyplot.show()
```

```python
fig2, ax = plt.pyplot.subplots()
ax.plot(combined['time'], combined['btc change/tether change'])

ax.set(xlabel='Time (hr)', ylabel='Hourly change in Bitcoin Price Divided by Hourly change in Tether Price (%)',
       title='Relative movements of Bitcoin Price with Respect to Tether')
ax.grid()
fig2.set_size_inches(30,24)
fig2.savefig("test.png")
plt.pyplot.show()
```

```python
r_corr = (combined['btc_change'].corr(combined['tether_change']))**2
print(r_corr)
```

```python
def factor_convert(x):
    x = float(x)
    if x>0:
        y=1
    elif x==0:
        y=0
    else:
        y=-1
    return y
   
combined['btc_direction']= combined['btc_change'].apply(factor_convert)
combined['tether_direction']= combined['tether_change'].apply(factor_convert)
combined
r_corr_dir = (combined['btc_direction'].corr(combined['tether_direction']))**2
print(r_corr_dir)
```

```python
matrix = pd.crosstab(combined['btc_direction'], combined['tether_direction'], margins = True)
print(matrix)
```

```python
def factor_convert_lim(x):
    x= float(x)
    if x>1:
        y=1
    elif x<-1:
        y=-1
    else:
        y=0
    return y
    
   
combined['btc_direction_limit']= combined['btc_change'].apply(factor_convert_lim)
combined['tether_direction_limit']= combined['tether_change'].apply(factor_convert_lim)
combined
r_corr_dir_limit = (combined['btc_direction_limit'].corr(combined['tether_direction_limit']))**2
print(r_corr_dir_limit)
matrix2= pd.crosstab(combined['btc_direction_limit'], combined['tether_direction'], margins = True)
print(matrix2)
```
